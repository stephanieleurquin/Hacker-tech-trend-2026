```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Chemin complet vers le CSV
csv_file = '/kaggle/input/datasets/kanchana1990/hacker-news-tech-trend-velocity-and-nlp/Hacker_News_Tech_Trend_Velocity_Final.csv'

# Lire le CSV
df = pd.read_csv(csv_file)

# Vérifier que ça fonctionne
df.head()

plt.figure(figsize=(8,6))
sns.scatterplot(x='Score_Velocity', y='Comment_Velocity', hue='Is_Viral', data=df)
plt.title("Score Velocity vs Comment Velocity")
plt.show()


# Histogramme du Score
plt.figure(figsize=(10,5))
sns.histplot(df['Score'], bins=50, kde=True)
plt.title("Distribution du Score des posts")
plt.xlabel("Score")
plt.ylabel("Nombre de posts")
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(x='Post_Type', hue='Is_Viral', data=df)
plt.title("Viralité selon le type de post")
plt.show()
```
